```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error 
from sklearn.datasets import load_iris 
```

```python
Iris=load_iris()
```

```python
Iris.feature_names
```

```python
Iris.keys()
```

```python
Iris_df = pd.DataFrame(Iris.data,columns=Iris.feature_names)
Iris_df["Species"]=Iris.target 
Iris_df.head() 
```

```python
 X= Iris_df.drop(['Species'],axis=1) 
 y= Iris_df['Species']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

```python
dt_classifier=DecisionTreeClassifier()
```

```python
dt_classifier.fit(X_train,y_train)
```

```python
dt_classifier.score(X_train,y_train)
```

```python
dt_classifier.score(X_test,y_test)
```

```python
y_pred=dt_classifier.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
print(mse)
```

```python
rms=(mse)**(1/2)
```

```python
rms
```

```python
from sklearn.tree import plot_tree
```

```python
from sklearn import tree 
plt.figure(figsize=(15,10)) 
tree.plot_tree(dt_classifier,filled=True) 
```

with HyperTuning

```python
dt_classifier = DecisionTreeClassifier(criterion="entropy",max_depth=5)
dt_classifier.fit(X_train,y_train)

score = dt_classifier.score(X_train,y_train)
print("R-squared Train  :", score)

score = dt_classifier.score(X_test, y_test)
print("R-squared Test  :", score)

y_pred = dt_classifier.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```
